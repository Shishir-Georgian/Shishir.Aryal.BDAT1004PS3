# Question No 1


1) Import the necessary libraries
2) Import the dataset from this address.
3) Assign it to a variable called users
4) Discover what is the mean age per occupation
5) Discover the Male ratio per occupation and sort it from the most to the least
6) For each occupation, calculate the minimum and maximum ages
7) For each combination of occupation and sex, calculate the mean age
8) For each occupation present the percentage of women and men

```python
# Step 1: Import the necessary libraries
import pandas as pd
from tabulate import tabulate

# Step 2: Import the dataset from this address
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

# Step 3: Assign it to a variable called users
users = pd.read_csv(url, sep='|', index_col = 'user_id')
```

```python
# Step 4: Discover what is the mean age per occupation

age = users[['occupation','age']].groupby("occupation").mean()
print(age)
```

```python
# Step 5: Discover the Male ratio per occupation and sort it from the most to the least

df = users[['occupation','gender']]
males = df.loc[df['gender']== 'M']
occupation = df.groupby('occupation').count()
male_count = males.groupby("occupation").count()
ratio = male_count/occupation
ratio.sort_values(ascending= False, by='gender')
```

```python
# Step 6: For each occupation, calculate the minimum and maximum ages

min_max_ages_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max']).reset_index()
print("\nMin and Max Ages per Occupation:\n", tabulate(min_max_ages_per_occupation, headers='keys', tablefmt='grid', showindex=False))
```

```python
# Step 7: For each combination of occupation and sex, calculate the mean age

mean_age_per_occupation_sex = users[['occupation','gender','age']].groupby(['occupation','gender']).mean('age')
mean_age_per_occupation_sex
```

```python
# Step 8: For each occupation present the percentage of women and men

gender_percentage_per_occupation = users[['occupation','gender','age']]
percentage = gender_percentage_per_occupation.groupby('occupation')['gender'].value_counts(normalize=True)*100
percentage
```

```python
# Step 1: Import the necessary libraries
import pandas as pd
from tabulate import tabulate

# Step 2: Import the dataset from the given address
url_path = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"

# Step 3: Assign it to a variable called euro12
euro12 = pd.read_csv(url_path)
```

```python
# Step 4: Select only the Goal column

goals = euro12['Goals']
print("\nGoals Column Selected\n")
print(tabulate(goals.to_frame(), headers='keys', tablefmt='pipe', showindex=False))
```

```python
# Step 5: How many teams participated in Euro 2012?

teams_in_Euro_2012 = euro12['Team'].nunique()
print("\nTeams Participated in Euro 2012:", teams_in_Euro_2012)
```

```python
# Step 6: What is the number of columns in the dataset?

total_columns = len(euro12.columns)
print("\nTotal numbers of columns in the Datasets are", total_columns)
```

```python
# Step 7: View only the columns Team, Yellow Cards, and Red Cards and assign them to a dataframe called discipline

discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print("\nDiscipline Dataframe Created\n")
print(tabulate(discipline, headers='keys', tablefmt='orgtbl', showindex=False))
```

```python
# Step 8: Sort the teams by Red Cards, then Yellow Cards

sorted_data = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)
print("\nSorted Data\n")
print(tabulate(sorted_data, headers='keys', tablefmt='fancy_grid', showindex=False))
```

```python
# Step 9: Calculate the mean Yellow Cards given per Team

mean_yellow_cards_per_team = discipline['Yellow Cards'].mean()
print("\nThe mean Yellow Cards per Team are", mean_yellow_cards_per_team)
```

```python
# Step 10: Filter teams that scored more than 6 goals

filter_teams_data = euro12[euro12['Goals'] > 6][['Team', 'Goals']]
print("\nteams with More Than 6 Goals\n")
print(tabulate(filter_teams_data, headers='keys', tablefmt='grid', showindex=False))
```

```python
# Step 11: Select the teams that start with G

starts_with_G = euro12[euro12['Team'].str.startswith('G')][['Team', 'Goals']]
print("\nTeams Starting with 'G' are below\n")
print(tabulate(starts_with_G, headers='keys', tablefmt='pretty', showindex=False))
```

```python
# Step 12: Select the first 7 columns

first_seven_columns = euro12.iloc[:, :7]
print("\nFirst Seven Columns are:\n")
print(tabulate(first_seven_columns, headers='keys', tablefmt='pretty', showindex=False))
```

```python
# Step 13: Select all columns except the last 3

except_last_three = euro12.iloc[:, :-3]
print("\nStep 13: All Columns Except the Last 3\n")
print(except_last_three)
```

```python
# Step 14: Present only the Shooting Accuracy from England, Italy, and Russia

shooting_data = euro12.loc[euro12['Team'].isin(['England', 'Italy', 'Russia']), ['Team', 'Shooting Accuracy']]
print("\nShooting Accuracy from England, Italy, and Russia are as follows:\n")
print(tabulate(shooting_data, headers='keys', tablefmt='pipe', showindex=False))
```

```python
# Step 5: Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns]. 

data.index = pd.to_datetime(pd.concat([data['Year'], data['Month'], data['Day']], axis=1))
data.drop(['Year','Month','Day'], axis=1, inplace=True)
data
```

```python
# Step 6: Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below. 

data.isna().sum()
```

```python
# entire record
data.isnull().sum().sum()
```

```python
#Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.

data.mean().mean()
```

```python
# Step 9:  Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days. A different set of numbers for each location.


stats = pd.concat([data.min(), data.max(), data.mean(), data.std()], axis=1) 
stats.rename(columns={0:'Min', 1:'Max', 2:'Mean', 3:'Std'}, inplace=True) 

stats
```

```python
# Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day. A different set of numbers for each day.

days = pd.concat([data.min(axis=1), data.max(axis=1), data.mean(axis=1), data.std(axis=1)], axis=1)
days.rename(columns={0:'Min',1:'Max',2:'Mean',3:'Std'}, inplace=True)
days
```

```python
# Step 14. Downsample the record to a weekly frequency for each location.

weekly = data.resample('W').ffill()

weekly
```

```python
import pandas as pd
from tabulate import tabulate

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

# Step 4: See the first 10 entries
print("\nStep 4: First 10 entries\n", tabulate(chipo.head(10), headers='keys', tablefmt='plain'))

# Step 5: Number of observations
num_observations = len(chipo)
print("\nStep 5: Number of observations are", num_observations)

# Step 6: Number of columns
num_columns = chipo.shape[1]
print("\nStep 6: Number of columns are", num_columns)

# Step 7: Print the name of all columns
column_names = chipo.columns.tolist()
print("\nStep 7: Columns\n", tabulate([(i,) for i in column_names], headers=['Column Name'], tablefmt='grid'))

# Step 8: How is the dataset indexed?
index_type = chipo.index
print("\nStep 8: Index Type =", type(index_type))

# Step 9: Most-ordered item
most_ordered_item = chipo['item_name'].mode().iloc[0]
print("\nStep 9: Most-ordered item is", most_ordered_item)

# Step 10: Quantity of the most-ordered item
quantity_most_ordered_item = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
print("\nStep 10: Quantity of most-ordered item is", quantity_most_ordered_item)

# Step 11: Most ordered item in the choice_description column
most_ordered_choice = chipo['choice_description'].mode().iloc[0]
print("\nStep 11: Most-ordered item in choice_description is", most_ordered_choice)

# Step 12: Total items ordered
total_items_ordered = chipo['quantity'].sum()
print("\nStep 12: Total items ordered are", total_items_ordered)

# Step 13: Turn item price into a float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
print("\nStep 13: Check item price type =", chipo['item_price'].dtype)

# Step 14: Revenue for the period
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("\nStep 14: Totl revenue is $",revenue)

# Step 15: Number of orders made
num_orders = chipo['order_id'].nunique()
print("\nStep 15: Total number of orders are", num_orders)

# Step 16: Average revenue amount per order
avg_revenue_per_order = revenue / num_orders
print("\nStep 16: The average amount per order is", avg_revenue_per_order)

# Step 17: Number of different items sold
num_different_items = chipo['item_name'].nunique()
print("\nStep 17: Total number of different items sold =", num_different_items)
```

# Question 7
Create a vertical bar chart comparing the number of marriages and divorces per 
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

data_path = "C:/Users/aryal/Desktop/Problem sets for data programming/us-marriages-divorces-1867-2014.csv"
data = pd.read_csv(data_path)

# Filter data for the years 1900, 1950, and 2000
years = [1900, 1950, 2000]
filtered_data = data[data['Year'].isin(years)].copy()

# Calculate per 1000 values using .loc indexer
filtered_data.loc[:, 'Marriages_per_1000'] = filtered_data['Marriages'] / filtered_data['Population']
filtered_data.loc[:, 'Divorces_per_1000'] = filtered_data['Divorces'] / filtered_data['Population']

# Create the vertical bar chart
plt.figure(figsize=(10, 4))

x = filtered_data['Year']
marriages_per_1000 = filtered_data['Marriages_per_1000']
divorces_per_1000 = filtered_data['Divorces_per_1000']

bar_width = 12
plt.bar(x - bar_width/2, marriages_per_1000, width=bar_width, label='Marriages per 1000',color = 'magenta')
plt.bar(x + bar_width/2, divorces_per_1000, width=bar_width, label='Divorces per 1000', color = 'green')

plt.xlabel('Year',color = 'red', fontsize = 18)
plt.ylabel('Per 1000',color = 'red', fontsize = 18)
plt.title('Comparison of Marriages and Divorces per 1000 in the U.S. (1900, 1950, 2000)',color = 'Black', fontsize = 18)
plt.xticks(years)
plt.legend()

plt.tight_layout()
plt.show()
```
